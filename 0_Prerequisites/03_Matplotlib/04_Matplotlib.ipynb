# Label Parameter

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.plot(days, max_t, label='max')
plt.plot(days, min_t, label='min')
plt.plot(days, avg_t, label='average')
```

## legend

```python
plt.plot(days, max_t, label='max')
plt.plot(days, min_t, label='min')
plt.plot(days, avg_t, label='average')

plt.legend(loc='best')
```

```python
plt.plot(days, max_t, label='max')
plt.plot(days, min_t, label='min')
plt.plot(days, avg_t, label='average')

plt.legend(loc='upper right', shadow=True, fontsize='large')
```

# Bar chart

```python
import numpy as np
```

```python
company = ['Google', 'Amazon', 'MS', 'Facebook']
revenue = [90, 136, 89, 27]
```

```python
xpos = np.arange(len(company)) # 0,1,2,3
xpos
```

```python
plt.bar(xpos, revenue, label='Revenue')
plt.ylabel("Revenue(Bln)")
plt.title("Technology Companies Stocks")
plt.legend()
```

```python
plt.bar(xpos, revenue, label='Revenue')

plt.xticks(xpos, company)

plt.ylabel("Revenue(Bln)")
plt.title("Technology Companies Stocks")
plt.legend()
```

```python
profit = [40,2,34,12]
```

```python
plt.bar(xpos, revenue, label='Revenue')
plt.bar(xpos, profit, label='Profit')

plt.xticks(xpos, company)

plt.ylabel("Revenue(Bln)")
plt.title("Technology Companies Stocks")
plt.legend()
```

```python
plt.bar(xpos - 0.2, revenue, label='Revenue')
plt.bar(xpos + 0.2, profit, label='Profit')

plt.xticks(xpos, company)

plt.ylabel("Revenue(Bln)")
plt.title("Technology Companies Stocks")
plt.legend()
```

```python
plt.bar(xpos - 0.2, revenue, width=0.4 , label='Revenue')
plt.bar(xpos + 0.2, profit, width=0.4 , label='Profit')

plt.xticks(xpos, company)

plt.ylabel("Revenue(Bln)")
plt.title("Technology Companies Stocks")
plt.legend()
```

# Pie chart

```python
value = [1400,600,300,410,250]
reason = ["Home Rent", "Food", "Internet", "Car", "Other"]
```

```python
plt.pie(value, labels = reason)
```

```python
plt.pie(value, labels = reason, autopct='%1.2f%%', radius=1.5)
```

```python
plt.pie(value, labels = reason, autopct='%.3f%%', radius=1.5)
```

# Histogram

```python
blood_sugar = [113, 85,90,150,149,88,93,115,135,80,77,82,129]
```

```python
plt.hist(blood_sugar, rwidth=0.8)
```

```python
plt.hist(blood_sugar, rwidth=0.8, bins = 4)
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number of Patients")
plt.title("Blood Sugar levels")

plt.hist(blood_sugar, bins=[80, 100] ,rwidth=0.9, color='g')
plt.hist(blood_sugar, bins=[100, 125] ,rwidth=0.9, color='b')
plt.hist(blood_sugar, bins=[125, 150] ,rwidth=0.9, color='r')
```
