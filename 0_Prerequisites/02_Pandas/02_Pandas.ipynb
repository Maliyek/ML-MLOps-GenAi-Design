# Handling Missing Values

```python
df = pd.read_csv('weather_data2.csv', parse_dates = ['day'])

df
```

```python
type(df['day'][0])
```

```python
df.set_index('day', inplace=True)
df
```

## fillna

fill not available

```python
new_df = df.fillna(0)

new_df
```

```python
new_df = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'No Event'
})
new_df
```

## forward fill

```python
new_df = df.fillna(method='ffill')
new_df
```

## backward fill

```python
new_df = df.fillna(method='bfill')
new_df
```

## column wise fill

```python
new_df = df.fillna(method='bfill', axis='columns')
new_df
```

```python
df
```

## dropna

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how = 'all')
new_df
```

```python
new_df = df.dropna(thresh=2)
new_df
```

## replace data

```python
df = pd.read_csv('weather_datamissing.csv')
df
```

```python
new_df = df.replace(-99999, value = np.NaN)
new_df
```

```python
new_df = df.replace([-99999, -88888], value = 0)
new_df
```

```python
new_df = df.replace({
    'temperature': -99999,
    'windspeed': -99999,
    'event': '0'
}, value=np.NaN)

new_df
```
